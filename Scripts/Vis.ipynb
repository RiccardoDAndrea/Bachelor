```python
import requests
import pandas as pd

# Definiere die URL zur Abfrage der Daten
url = 'http://api.worldbank.org/v2/country/GR/indicator/NY.GDP.PCAP.CD?format=json'

# Fhre die Abfrage durch
response = requests.get(url)

# berprfe, ob die Abfrage erfolgreich war (Statuscode 200)
if response.status_code == 200:
    # Extrahiere die Daten im JSON-Format
    data = response.json()
    
    # Die eigentlichen Daten sind im zweiten Element der Liste
    gdp_data = data[1]
    
    # Konvertiere die Daten in ein DataFrame
    greece_gdp = pd.DataFrame(gdp_data)
    
    # Benenne die Spalten um und entferne unntige Spalten
    greece_gdp.rename(columns={'value': 'GDP per Capita Greece', 'date':'Date'}, inplace=True)
    greece_gdp.drop(columns=['indicator', 'country', 'countryiso3code', 'unit', 'obs_status', 'decimal'], inplace=True)

    # Filtere das DataFrame nach dem gewnschten Zeitraum (1995 bis 2022)
    greece_gdp = greece_gdp.query('Date >= "1995" & Date <= "2022"')
    
    # Zeige das gefilterte DataFrame an
    greece_gdp
else:
    print('Fehler beim Abrufen der Daten:', response.status_code)
```
